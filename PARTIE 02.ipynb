```python
import cv2
import numpy as np
from keras.models import load_model
from keras.preprocessing import image
from keras.applications.mobilenet_v2 import preprocess_input

#Lien vers le modele kaggle : https://www.kaggle.com/datasets/abhisheksingh016/machine-model-for-emotion-detection/code
model = load_model('face_model.h5')  


# Les émotions détectées
emotion_labels = ['Angry', 'Disgusted', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# Démarrer la webcam
cap = cv2.VideoCapture(0)

# Charger le classificateur en cascade pour la détection de visages
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convertir l'image en niveaux de gris
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Détecter les visages dans l'image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Traiter chaque visage détecté
    for (x, y, w, h) in faces:
        # Extraire le visage et le redimensionner
        face = frame[y:y+h, x:x+w]

        # Prétraiter l'image pour le modèle
        
        # face_img = image.img_to_array(face_resized)
        # face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)
        # face_img = np.expand_dims(face_img, axis=0)
        # face_img = preprocess_input(face_img)

        face_img = cv2.resize(face, (48, 48))#dimention attendu par le modéle
        face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)# convertir face_img en gris attentes du modèle
        face_img = image.img_to_array(face_img)
        face_img = np.expand_dims(face_img, axis=0)
        face_img = np.vstack([face_img])# convertir face_img en une forme compatible avec les attentes du modèle
        # Prédire l'émotion
        emotion = model.predict(face_img)

        # Afficher l'émotion détectée
        emotion_index = np.argmax(emotion)
        emotion_label = emotion_labels[emotion_index]
        
        # Afficher le label sur l'image
        cv2.putText(frame, emotion_label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Afficher l'image
    cv2.imshow('Emotion Detection', frame)

    # Quitter si la touche 'q' est pressée
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter('test_emotion.avi', fourcc, 20.0, (640, 480))
    out.write(frame)

cap.release()
cv2.destroyAllWindows()
```
